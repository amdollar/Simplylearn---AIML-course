# Sequential Chains
Output of one chain is the Input to Another chain

```python
from dotenv import find_dotenv, load_dotenv
load_dotenv(find_dotenv())
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatOpenAI(model = 'gpt-4o-mini', temperature=0.2)
```

```python
system_message_persona = SystemMessagePromptTemplate.from_template('You are {school} teacher. You answer in short sentences. Make the students life difficult to understand the concept')
human_message_query = HumanMessagePromptTemplate.from_template('Write {lines}, about {topic}')

message = [system_message_persona, human_message_query]

template = ChatPromptTemplate(message)

teacher_chain = template | llm | StrOutputParser()
```

```python
evaluation_prompt = ChatPromptTemplate.from_template('''Analyse the following text {response} you need to tell me how difficult it is to understand Answer in one sentence only''')

evaluation_prompt_chain = evaluation_prompt | llm | StrOutputParser()
```

```python
final_chain = {'response': teacher_chain} | evaluation_prompt_chain

output = final_chain.invoke({
    'school':'PHD',
    'lines': '3',
    'topic': 'Linear Algebra',
})
```
