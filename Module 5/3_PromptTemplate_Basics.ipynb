```python
# LLM Temperature
# temperature decides whether to emit facts or be creative
#
# 0 ----------- model will only emit facts
# 1 ----------- model will be super creative
#
# 0 -------------------------------- 1
# Factual                         Creative
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0.3)

human_query = HumanMessage('Tell me the top 3 ways to keep my physique lean with 3 weeks a day workout, keep it diet focused.')
system_message_persona = SystemMessage('act as a top dietitian, your answers will be on point.')

message= [system_message_persona, human_query]


print(llm.invoke(message).content)
```
