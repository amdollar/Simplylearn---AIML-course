```python
from dotenv import find_dotenv, load_dotenv

load_dotenv(find_dotenv())
```

```python
from langchain_openai import ChatOpenAI

from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, PromptTemplate, ChatPromptTemplate
```

```python
llm = ChatOpenAI(model = 'gpt-4o-mini', temperature = 0.2)
```

```python
system_message_persona = SystemMessagePromptTemplate.from_template('Act as a {persona}, your answers are short sentences in {} language.')
human_message_input = HumanMessagePromptTemplate.from_template('Tell me about {topic} in {points} points.')

message = [system_message_persona, human_message_input]

template = ChatPromptTemplate(message)

# Simple Chain:
# A simple chain consist of a sequence of messages that are passed to the LLm Model in a specific order,
# In this case human message is followed by system message in the sequence

chain = template | llm

print(type(chain))

```
