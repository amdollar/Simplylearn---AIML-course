```python
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())
```

```python
from langchain_openai import ChatOpenAI

from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate
from langchain_core.runnables import RunnableParallel
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatOpenAI(model = 'gpt-4o-mini', temperature=0.4)
```

```python

# creation of one chain:

system_message = SystemMessagePromptTemplate.from_template('Act as a {persona}, your answers are short and concise.')
human_message = HumanMessagePromptTemplate.from_template('What is the best way of learning {topic}, tell me in {points} points.')

message = [system_message, human_message]

template = ChatPromptTemplate(message)

chain1 = template | llm | StrOutputParser()
```

```python
system_message_2 = SystemMessagePromptTemplate.from_template('Act as a {persona}.')
human_message2 = HumanMessagePromptTemplate.from_template('Give me {count} posts for social media platform: {platform}, on {topic}, keep the tone logical and funny.')

message2 = [system_message_2, human_message2]


template = ChatPromptTemplate(message2)

chain2 = template | llm | StrOutputParser()
```

```python
# class to build the final chain:
finalchain = RunnableParallel(facts = chain1, posts = chain2)

output = finalchain.invoke({
    'persona':'Boxing Champion',
    'topic': 'Shadow Boxing',
    'points':'4',
    'persona':'Top Social Media Manager',
    'count':'4',
    'platform':'X/Twitter',
    'topic':'Vibe coding'
})

```
