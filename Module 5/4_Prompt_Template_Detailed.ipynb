```python
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import (
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
    ChatPromptTemplate
)
```

```python
# Test code to show the working of placeholders

system= SystemMessagePromptTemplate.from_template('Act as a {persona} professor. Your answers are in short Sentences.')
question = HumanMessagePromptTemplate.from_template('Tell me about {topic} in {points} points')

system.format(persona='math')
```

```python

question.format(topic= 'Linear algebra', points=3)
```

```python
llm = ChatOpenAI(model = 'gpt-4o-mini', temperature = 0.2)

system= SystemMessagePromptTemplate.from_template('Act as a {persona} professor. Your answers are in short Sentences.')
question = HumanMessagePromptTemplate.from_template('Tell me about {topic} in {points} points')


message = [system, question]

templates = ChatPromptTemplate(message)

questions1List = templates.invoke({
    'persona':'math',
    'topic':'linear algebra',
    'points': 2
})

questions1List
```

```python
questions2List= templates.invoke({
    'persona':'AI',
    'topic': 'Transformers',
    'points': 3
})
questions2List


# This is how we can create multiple questions list (Request body to the LLM) By defining one common System 
# and Human Messages.
```

```python
print(llm.invoke(questions1List).content)
```

```python
print(llm.invoke(questions2List).content)
```
