```python
# 1. Load the files using Directory loader 
# 2. Read these files and make document
# 3. Create chunks for this document
# 4. Create embeddings for these chunks
# 5. Feed this embedded data to the VectorDatabase
# 6. Feed this Vector database as a RAG storage to the LLM
# 7. Use llm to fetch the response and get the queries answered


from langchain_community.document_loaders import TextLoader, DirectoryLoader
from langchain_classic.text_splitter import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
# from langchain.retrievers import EnsembleRetriever
from langchain_classic.retrievers import BM25Retriever, EnsembleRetriever
# from langchain_community.retrievers import BM25Retriever

dir = DirectoryLoader(path = './Dataset', loader_cls=TextLoader)
documents = dir.load()

```

```python
# 2. Create chunks using RecursiveCharacterTextSplitter class

splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
chunks = splitter.split_documents(documents)
len(chunks)
```

```python
# 3. Create embeddings from these Chunks using OpenAIEmbeddings:
embeddings= OpenAIEmbeddings(model='text-embedding-3-small')
chunk_list = [chunk.page_content for chunk in chunks]

embedded_chunks = embeddings.embed_documents(chunk_list)

print(f'Shape of embedded_chunks : {len(embedded_chunks)}')
```

```python
# 4. Feed these embedded data to the VectorStorage

storage = FAISS.from_documents(chunks, embeddings)

res = storage.similarity_search('What is the buffet breakfast timings in our hotel ? ', k=2)

for doc in res:
    print(doc.page_content)


# Initialize the Retriever: The retriever is typically an interface around your vector store that you can configure with specific parameters, such as the number of documents to fetch (top-k).

```

```python
# Next task is to build domain specific retrievers:
# 1. Setup your keywoard retriever: (BM25)

bm25_retriever =   BM25Retriever.from_documents(chunks)
bm25_retriever.k = 3

# 2. Setup sementic retriever:
retriever = storage.as_retriever(search_kwargs={'k': 3})

# 3. Create Domain Specific Ensemble:

ensemble_retriever = EnsembleRetriever(
    retrievers=[retriever, bm25_retriever],
    weights = [0.5, 0.5] 
)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_classic.chains.retrieval import create_retrieval_chain

model = ChatOpenAI(model= 'gpt-4o-mini')

prompt = '''You are expert in Answering general queries from the customers
{context}
Question: {input}
Answer:
'''

prompt_template = PromptTemplate(template = prompt, input_variables=['context','input'])
stuff_document_chain = create_stuff_documents_chain(model, prompt_template)

retriever_chain = create_retrieval_chain(ensemble_retriever, stuff_document_chain)

response = retriever_chain.invoke({'input':'Tell me about hotel policies'})

response['answer']

```
