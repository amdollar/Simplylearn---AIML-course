```python
# 1. Load the files using Directory loader 
# 2. Read these files and make document
# 3. Create chunks for this document
# 4. Create embeddings for these chunks
# 5. Feed this embedded data to the VectorDatabase
# 6. Feed this Vector database as a RAG storage to the LLM
# 7. Use llm to fetch the response and get the queries answered


from langchain_community.document_loaders import TextLoader, DirectoryLoader
from langchain_classic.text_splitter import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings

dir = DirectoryLoader(path = './Dataset', loader_cls=TextLoader)
documents = dir.load()

```

```python
# 2. Create chunks using RecursiveCharacterTextSplitter class

splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
chunks = splitter.split_documents(documents)
len(chunks)
```

```python
# 3. Create embeddings from these Chunks using OpenAIEmbeddings:
embeddings= OpenAIEmbeddings(model='text-embedding-3-small')
chunk_list = [chunk.page_content for chunk in chunks]

embedded_chunks = embeddings.embed_documents(chunk_list)

print(f'Shape of embedded_chunks : {len(embedded_chunks)}')
```
