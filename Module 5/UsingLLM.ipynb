```python
# Method: Streaming response (useful for long responses)
def generate_streaming(prompt):
    """Generate streaming response - prints text as it's generated"""
    response = model.generate_content(prompt, stream=True)
    full_text = ""
    for chunk in response:
        chunk_text = chunk.text
        print(chunk_text, end="", flush=True)
        full_text += chunk_text
    return full_text

# Example usage
streaming_result = generate_streaming("Explain machine learning in 3 paragraphs")
```

```python
# Method: Chat conversation (maintains context)
def start_chat():
    """Start a chat conversation that maintains context"""
    chat = model.start_chat(history=[])
    return chat

def send_message(chat, message):
    """Send a message in an ongoing chat"""
    response = chat.send_message(message)
    return response.text

# Example usage
# chat = start_chat()
# response1 = send_message(chat, "My name is John")
# response2 = send_message(chat, "What's my name?")  # Chat remembers context
# print(response2)
```

```python
# Method: Generate with custom parameters
def generate_with_params(prompt, temperature=0.7, max_tokens=1024):
    """
    Generate text with custom parameters
    
    Args:
        prompt: Input text prompt
        temperature: Controls randomness (0.0-1.0), lower = more focused
        max_tokens: Maximum length of response
    """
    generation_config = {
        "temperature": temperature,
        "max_output_tokens": max_tokens,
    }
    response = model.generate_content(prompt, generation_config=generation_config)
    return response.text

# Example usage - creative vs focused
# creative = generate_with_params("Write a story", temperature=0.9)  # More creative
# focused = generate_with_params("Explain quantum physics", temperature=0.3)  # More focused
```
