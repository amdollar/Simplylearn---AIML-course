```python
import numpy as np
import tensorflow as tf
```

```python
model = tf.keras.applications.VGG19()
```

```python
model.summary()
```

```python
image = tf.keras.preprocessing.image.load_img('African_Bush_Elephant.jpg', target_size=(224,224))
img_arr = tf.keras.preprocessing.image.img_to_array(image)

img_arr.shape

np_img = np.expand_dims(img_arr, axis = 0)
np_img.shape
```

```python
prediction_probability = model.predict(np_img)

prediction_val = tf.keras.applications.vgg19.decode_predictions(prediction_probability, top = 2)

object_names = [name for _, name, _ in prediction_val[0]]


object_names
```
