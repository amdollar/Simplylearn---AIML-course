```python
import numpy as np
import tensorflow as tf
```

```python
model = tf.keras.applications.Xception()
model.summary()
```

```python
image = tf.keras.preprocessing.image.load_img('African_Bush_Elephant.jpg', target_size=(299, 299))
image_arr = tf.keras.preprocessing.image.img_to_array(image)

image_arr.shape
```

```python
np_image= np.expand_dims(image_arr, axis = 0)
np_image.shape
```

```python
probabilty_prediction = model.predict(np_image)

probabilty_prediction_top2 = tf.keras.applications.xception.decode_predictions(probabilty_prediction, top=2)


obj_names = [name for _, name, _ in probabilty_prediction_top2[0]]
obj_names
```
