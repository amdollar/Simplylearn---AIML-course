```python
import numpy as np
import tensorflow as tf
```

```python
model = tf.keras.applications.MobileNet()
```

```python
model.summary()
```

```python
image = tf.keras.preprocessing.image.load_img('African_Bush_Elephant.jpg', target_size = (224,224))
image_arr = tf.keras.preprocessing.image.img_to_array(image)

image_arr.shape

# Shape expansion for compatibility

np_image_arr = np.expand_dims(image_arr, axis= 0)

np_image_arr.shape
```

```python
probability_prediction = model.predict(np_image_arr)

top_2_probability_prediction = tf.keras.applications.mobilenet.decode_predictions(probability_prediction)

object_names = [name for _, name, _ in top_2_probability_prediction[0]]


object_names
```
