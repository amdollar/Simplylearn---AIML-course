```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.applications.vgg16.VGG16()
```

```python
model.summary()
```

```python
image = tf.keras.preprocessing.image.load_img(path='African_Bush_Elephant.jpg', target_size=(224,224))

image_arr = tf.keras.preprocessing.image.img_to_array(image)

image_arr.shape
```

```python
image_arr_process = np.expand_dims(image_arr, axis=0)

image_arr_process.shape
```

```python
prediction = model.predict(image_arr_process)

prediction_prob_arr = tf.keras.applications.vgg16.decode_predictions(prediction, top = 2)
elephant_names = [name for _, name, _ in prediction_prob_arr[0]]
elephant_names
```
